```python
#@title **Clone Files / Folders with GClone**
#@markdown > *To clone, make sure links / IDs are set to public or in a way that your Service Accounts have access to them. If your Service Accounts don't have access, then you cannot clone!*

#@markdown > **Below, type in the source and destination GDrive folder IDs for your clone operation.**

#@markdown > **`$src` is the source file / folder ID that you want to clone and `$dst` is where the destination folder of what you are cloning is.**

# Source and Destination ID Variables
src = '' #@param {type:"string"}
dst = '' #@param {type:"string"}

# GClone Copy Command
!gclone copy GC:{$src} GC:{$dst} --transfers 50 -vP --stats-one-line --stats=15s --ignore-existing --drive-server-side-across-configs --drive-chunk-size 128M --drive-acknowledge-abuse --drive-keep-revision-forever
!clear
print("||----- Cloning Complete! -----||")
```

```python
#@title **Move Files / Folders with GClone**
#@markdown > *To move files / folders, make sure your Service Accounts have permissions to view, and move around things in your drive or else you cannot run this command.*

#@markdown > **Below, type in the source and destination GDrive folder IDs for your move operation.**

#@markdown > **`$src` is the source file / folder ID that you want to move and `$dst` is where the destination folder of what you are moving.**

# Source and Destination ID Variables
src = '' #@param {type:"string"}
dst = '' #@param {type:"string"}

# GClone Move Command
!gclone move GC:{$src} GC:{$dst} --transfers 50 --tpslimit-burst 50 --checkers 10 -vP --stats-one-line --stats=15s --ignore-existing --drive-server-side-across-configs --drive-chunk-size 128M --drive-acknowledge-abuse --drive-keep-revision-forever --fast-list
!clear
print("||----- Transfer Complete! -----||")
```
